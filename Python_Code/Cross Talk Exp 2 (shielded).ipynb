```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

All imported data:

```python
CT5data=pd.read_csv('CrossTalk5.csv')   #GHz shielded data
CT3data=pd.read_csv('CrossTalk3.csv')   #GHz range, unshielded data
CT6data=pd.read_csv('CrossTalk6.csv')   #MHz shielded data
CT4data=pd.read_csv('CrossTalk4.csv')   #MHz range, unshielded data
Floor3data=pd.read_csv('Floor3.csv')    #MHz range, no equipment connected
Floor4data=pd.read_csv('Floor4.csv')    #GHz range, no equipment connected
TFdata=pd.read_csv('Transformer1.csv')  #GHz range, only transformer
TF2data=pd.read_csv('Transformer2.csv') #MHz range, only transformer
```

## Cross Talk between 36 and shielded 32 (5GHz range)

```python
plt.figure(figsize=(16,8))
plt.scatter((10.**-9)*CT5data['Frequency'],CT5data['Attenuation'])
plt.xlabel('Frequency [GHz]')
plt.ylabel('Attenuation [dB]')
plt.title('Attenuation vs. Frequency for crosstalk between 32 shielded and 36 AWG')
plt.show()
```

Plotted above is the baseline recorded data for 32 AWG shielded twisted pair and 36 AWG twisted pair high voltage wires. 

```python
plt.figure(figsize=(16,8))
plt.scatter((10.**-9)*CT5data['Frequency'],CT5data['Attenuation'], label='32 Shielded')
plt.scatter((10.**-9)*CT3data['Frequency'],CT3data['Attenuation'], label='32 Unshielded')
plt.xlabel('Frequency [GHz]')
plt.ylabel('Attenuation [dB]')
plt.title('Shielded vs. Unshielded 32 AWG')
plt.legend()
plt.show()
```

Above is a comparison between the shielded and unshielded 32 AWG data taken

Below is the same plot with corrections for attenuation of within the transformer driving each 32 AWG leg

```python
plt.figure(figsize=(16,8))
plt.scatter((10.**-9)*CT5data['Frequency'],CT5data['Attenuation']-TFdata['Attenuation'], label='32 Shielded')
plt.scatter((10.**-9)*CT3data['Frequency'],CT3data['Attenuation']-TFdata['Attenuation'], label='32 Unshielded')
plt.xlabel('Frequency [GHz]')
plt.ylabel('Attenuation [dB]')
plt.title('Shielded vs. Unshielded 32 AWG, adjusted')
plt.legend()
plt.show()
```
